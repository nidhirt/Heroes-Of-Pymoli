```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head()
```

## Player Count

* Display the total number of players


```python
purchase_data_count = purchase_data["SN"].value_counts()
#print(purchase_data_count)

purchase_data_count_df = pd.DataFrame({"Total Players " : [len(purchase_data_count)] })

purchase_data_count_df

```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python


#get number of unique items
unique_items = purchase_data["Item ID"].value_counts()

#get average price 
average_price = purchase_data["Price"].mean()

#get number of purchases
purchase_count = purchase_data["Purchase ID"].value_counts()

#get total revenue
total_revenue = purchase_data["Price"].sum()

#create a dataFrame
summary_table = pd.DataFrame({
                    "Number of Unique Items" : [len(unique_items)],
                     "Average Price" :    average_price,
                     "Number of Purchases" :[len(purchase_count)],
                     "Total Revenue"  :   total_revenue})
summary_table["Average Price"] = summary_table["Average Price"].map("${:.2f}".format)

summary_table["Total Revenue"] = summary_table["Total Revenue"].map("${:,.2f}".format)

summary_table

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python

#drop duplicates from SN
purchase_data_notnull = pd.DataFrame(purchase_data.drop_duplicates(["SN"]))

#Group by Gender
purchase_data_gender = purchase_data_notnull.groupby(["Gender"])

#Total players count
Total_players = purchase_data_notnull["SN"].nunique()

purchase_percent = pd.DataFrame({ "Total Count" : purchase_data_gender["Gender"].count(),
                                "Percentage of Players" : purchase_data_gender["Gender"].count()/Total_players *100
                                })
purchase_percent["Percentage of Players"] = purchase_percent["Percentage of Players"].map("{:.2f}".format)

purchase_percent_sorted = purchase_percent.sort_values(["Percentage of Players"], ascending=False)


#purchase_percent_sorted.style.set_properties(**{'text-align': 'right'})

purchase_percent_sorted
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

#total_players = len(purchase_data["SN"].value_counts())

#group by gender

purchase_data_df = purchase_data.groupby(["Gender"])

#do all calculations
purchase_count = purchase_data_df["Purchase ID"].count()

average_purchase_price = purchase_data_df["Price"].mean()

total_purchase_value = purchase_data_df["Price"].sum()

total_gender_count = purchase_data_df["SN"].nunique()

avg_purchase_per_person = total_purchase_value/total_gender_count



purchase_summary_df = pd.DataFrame({"Purchase Count"  : purchase_count,
                                         "Average Purchase Price": average_purchase_price,
                                         "Total Purchase Value"  : total_purchase_value,
                                         "Average Total Purchase per Person" : avg_purchase_per_person
                                        
})


purchase_summary_df["Average Purchase Price"] = purchase_summary_df["Average Purchase Price"].map("${:.2f}".format)
purchase_summary_df["Total Purchase Value"] = purchase_summary_df["Total Purchase Value"].map("${:,.2f}".format)
purchase_summary_df["Average Total Purchase per Person"] = purchase_summary_df["Average Total Purchase per Person"].map("${:.2f}".format)

purchase_summary_df

```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#create bins

bins= [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names=["<10","10-14","15-19","20-24","25-29","30-34", "35-39","40+"]

purchase_data["Age Group"] =pd.cut(purchase_data["Age"],bins,labels=group_names)

player_grouped = purchase_data.groupby("Age Group")

#Count of players by age bins
player_count = player_grouped["SN"].nunique()

#% of players bin wise

percentage_of_players = (player_count/Total_players)*100

player_summary_df = pd.DataFrame({
                                    "Total Count" : player_count,
                                    "Percentage of Players" : percentage_of_players
                                })


player_summary_df["Percentage of Players"] = player_summary_df["Percentage of Players"].map("{:.2f}".format)

player_summary_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

#Retrieve the Item ID , Name and price columns

items = purchase_data[["Item ID", "Item Name", "Price"]]

# Group the item data by item id and item name 
item_stats = items.groupby(["Item ID","Item Name"])

# Count the number of times an item has been purchased 
purchase_count_item = item_stats["Price"].count()

# Purchase value per item 
purchase_value = (item_stats["Price"].sum()) 

# individual item price
item_price = purchase_value/purchase_count_item

# Create data frame with obtained values
most_popular_items = pd.DataFrame({"Purchase Count": purchase_count_item, 
                                   "Item Price": item_price,
                                   "Total Purchase Value":purchase_value})

# Sort in descending order to obtain top spender names and provide top 5 item names
popular_formatted = most_popular_items.sort_values(["Purchase Count"], ascending=False).head()


popular_formatted["Item Price"] = popular_formatted["Item Price"].map("${:.2f}".format)

popular_formatted["Total Purchase Value"] = popular_formatted["Total Purchase Value"].map("${:.2f}".format)


popular_formatted
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order

* Display a preview of the data frame



```python

# sort by purchase value in descending order

popular_bypurchasevalue= most_popular_items.sort_values(["Total Purchase Value"],ascending=False).head()

popular_bypurchasevalue["Item Price"] = popular_bypurchasevalue["Item Price"].map("${:.2f}".format)

popular_bypurchasevalue["Total Purchase Value"] = popular_bypurchasevalue["Total Purchase Value"].map("${:.2f}".format)


popular_bypurchasevalue
```
